*4.3.2 –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π*

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / (np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x)))

x = np.linspace(1, 100, 100)

y = my_func(x)

x_slice = x[:50]
y_slice = y[:50]

plt.plot(x, y, label='–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤ (–ª–∏–Ω–µ–π–Ω—ã–π)')

plt.scatter(x_slice, y_slice, c='green', label='–°—Ä–µ–∑ (—Ç–æ—á–µ—á–Ω—ã–π)')

plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏')
plt.legend()

plt.show()
```

*4.3.3 –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å
–ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(),
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ
–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz*

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 30, 1)

y = my_func(x)

plt.fill_between(x, y, alpha=0.5)
plt.plot(x, y, label='|cos(xe^(cos(x)+ln(x+1)))|')

area = np.trapz(y, x)

plt.text(5, 0.5, f'–ü–ª–æ—â–∞–¥—å = {area:.2f}', fontsize=12, ha='center')

plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏')
plt.legend()

plt.show()

print(f'–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º: {area:.2f}')
```

–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^(ùë•+ùë¶), sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^ùë¶.*

```python
import math

def main():
    print("–ü—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä")
    print("–î–æ—Å—Ç—É–ø–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:")
    print("1. –°–ª–æ–∂–µ–Ω–∏–µ (+)")
    print("2. –í—ã—á–∏—Ç–∞–Ω–∏–µ (-)")
    print("3. –£–º–Ω–æ–∂–µ–Ω–∏–µ (*)")
    print("4. –î–µ–ª–µ–Ω–∏–µ (/)")
    print("5. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ e^(x+y)")
    print("6. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ sin(x+y)")
    print("7. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ cos(x+y)")
    print("8. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ x^y")

    choice = input("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –æ–ø–µ—Ä–∞—Ü–∏–∏ (1/2/3/4/5/6/7/8): ")

    if choice in ['1', '2', '3', '4', '5', '6', '7', '8']:
        x = float(input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ (x): "))
        y = float(input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ (y): "))

        if choice == '1':
            result = x + y
        elif choice == '2':
            result = x - y
        elif choice == '3':
            result = x * y
        elif choice == '4':
            if y != 0:
                result = x / y
            else:
                result = "–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å!"
        elif choice == '5':
            result = math.exp(x + y)
        elif choice == '6':
            result = math.sin(x + y)
        elif choice == '7':
            result = math.cos(x + y)
        elif choice == '8':
            result = x ** y

        print("–†–µ–∑—É–ª—å—Ç–∞—Ç: ", result)
    else:
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º—ã–π –≤—ã–±–æ—Ä –æ–ø–µ—Ä–∞—Ü–∏–∏.")

if __name__ == "__main__":
    main()
```
