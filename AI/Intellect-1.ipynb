*4.3.2 Дана функция. Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный*

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / (np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x)))

x = np.linspace(1, 100, 100)

y = my_func(x)

x_slice = x[:50]
y_slice = y[:50]

plt.plot(x, y, label='Основной массив (линейный)')

plt.scatter(x_slice, y_slice, c='green', label='Срез (точечный)')

plt.xlabel('x')
plt.ylabel('Значение функции')
plt.legend()

plt.show()
```

*4.3.3 Дана функция. Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь
под ним. Для вычисления площади воспользуйте функции trapz(),
вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие
библиотеки:
from scipy.integrate import simps
from numpy import trapz*

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 30, 1)

y = my_func(x)

plt.fill_between(x, y, alpha=0.5)
plt.plot(x, y, label='|cos(xe^(cos(x)+ln(x+1)))|')

area = np.trapz(y, x)

plt.text(5, 0.5, f'Площадь = {area:.2f}', fontsize=12, ha='center')

plt.xlabel('x')
plt.ylabel('Значение функции')
plt.legend()

plt.show()

print(f'Площадь под графиком: {area:.2f}')
```

*4.3.5 Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: 𝑒^(𝑥+𝑦), sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥^𝑦.*

```python
import math

def main():
    print("Простейший калькулятор")
    print("Доступные операции:")
    print("1. Сложение (+)")
    print("2. Вычитание (-)")
    print("3. Умножение (*)")
    print("4. Деление (/)")
    print("5. Вычисление e^(x+y)")
    print("6. Вычисление sin(x+y)")
    print("7. Вычисление cos(x+y)")
    print("8. Вычисление x^y")

    choice = input("Введите номер операции (1/2/3/4/5/6/7/8): ")

    if choice in ['1', '2', '3', '4', '5', '6', '7', '8']:
        x = float(input("Введите первое число (x): "))
        y = float(input("Введите второе число (y): "))

        if choice == '1':
            result = x + y
        elif choice == '2':
            result = x - y
        elif choice == '3':
            result = x * y
        elif choice == '4':
            if y != 0:
                result = x / y
            else:
                result = "Деление на ноль!"
        elif choice == '5':
            result = math.exp(x + y)
        elif choice == '6':
            result = math.sin(x + y)
        elif choice == '7':
            result = math.cos(x + y)
        elif choice == '8':
            result = x ** y

        print("Результат: ", result)
    else:
        print("Недопустимый выбор операции.")

if __name__ == "__main__":
    main()
```
